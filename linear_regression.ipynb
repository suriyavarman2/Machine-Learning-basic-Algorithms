```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
data=pd.read_csv('winequality-red.csv')
data.plot()
df1 = pd.DataFrame(data,columns=['volatile acidity','citric acid'])
df1.head
```

```python
data=pd.read_csv('veg.csv')
data.plot()
df1 = pd.DataFrame(data,columns=['Nr. reviews','Nr. hotel reviews'])
df1.head
```

```python
plt.scatter(df1['volatile acidity'], df1['citric acid'], color='green')
plt.title('Regression among pH and residual sugar')
plt.xlabel('X - axis   - pH')
plt.ylabel('Y- Dependent - residual sugar')
```

```python
plt.scatter(df1['Nr. reviews'], df1['Nr. hotel reviews'], color='green')
plt.title('Regression among Nr. reviews and Nr. hotel reviews')
plt.xlabel('X - axis   - Nr. reviews')
plt.ylabel('Y- Dependent - Nr. reviews')
```

```python
week = df1['volatile acidity'].values.reshape(1,-1)
sales = df1['citric acid'].values.reshape(1,-1)
```

```python
X = df1[['volatile acidity']] 
y = df1['citric acid']
regr = linear_model.LinearRegression()
regr.fit(X,y)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
print('\nThe Regression Equation is',regr.coef_,'* X+',regr.intercept_)
```

```python

X = df1[['Nr. reviews']] 
y = df1['Nr. hotel reviews']
regr = linear_model.LinearRegression()
regr.fit(X,y)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
print('\nThe Regression Equation is',regr.coef_,'* X+',regr.intercept_)
```

```python
pred = regr.predict(X)
plt.scatter(df1['volatile acidity'], df1['citric acid'], color='green')
plt.plot(X,pred)
```

```python
pred = regr.predict(X)
plt.scatter(df1['Nr. reviews'], df1['Nr. hotel reviews'], color='green')
plt.plot(X,pred)
```

```python
print("\nAdjusted R Squared for Regression model:",regr.score(X,y))
```
